## AgriBrain: AI-Powered Agricultural Market Intelligence System.

AgriBrain is an end-to-end agritech recommendation engine that leverages real-time soil, weather, and market data to optimize crop selection for Kenyan farmers. The platform actively combats the agricultural "Cobweb Phenomenon"- a cycle of market oversupply and price crashes caused by farmer reliance in past planted crops and historical prices as their main source of choice. By utilizing predictive machibe learning and an LLM-driven market analysis agent, AgriBrain breaks this cycle by recommending high-demand, ecologically viable viable alternative crops that maximize harvest profitability.





## Key Stakeholders
1. Farmers(Primary users)
Need guidance on **what to grow** and **when to grow**

This will help them **avoid overproduction** and **maximize profits**

2. Government and Policy Makers
Use insights for **food security planning** and **import and export decision** and also **issue early warnings for shortages**


## What our project will solve.
The current project will help solve the current challenges that the farmers are facing such as **relying in guesswork** and **use of past experience** which often leads to oversupply and price crashes and shortages which leads to food insecurity. **This predictve intelligence will combine different data and use machine learning to guide farmers in their agricultural decisions**

Step 3: Normalizing the columns mapping before mapping
- Changing our columns to lowercase and removing whitespace will data inconsistencies.
- Mapping is also case-sensitive. If there is a slight mismatch in our maping dictionary, there won't be any matching and the mapping might return NaN.

```python
def normalize_item_name(x):
    x = str(x).lower().strip()
    x = x.replace(" and products", "")
    x = x.replace(", other", "")
    return x.strip()
```

```python
price_coverage = master_df.groupby("crop_name")["price_lcu_per_tonne"].agg(
    total_rows="size",
    non_null_prices=lambda x: x.notnull().sum(),
    percent_available=lambda x: x.notnull().mean() * 100
).reset_index()

print(price_coverage.sort_values("percent_available"))
```

Step: Filter dataset properly

```python
good_crops = price_coverage[
    price_coverage["percent_available"] >= 50
]["crop_name"].tolist()

master_model_df = master_df[
    master_df["crop_name"].isin(good_crops)
].copy()

print("final crops:", len(good_crops))
print(sorted(good_crops))

print("new shape:", master_model_df.shape)
print("remaining missing prices:", master_model_df["price_lcu_per_tonne"].isnull().sum())
```

Next step: interpolate prices within each crop

```python
master_model_df = master_model_df.sort_values(["crop_name", "year"]).copy()

master_model_df["price_lcu_per_tonne"] = master_model_df.groupby("crop_name")[
    "price_lcu_per_tonne"
].transform(lambda x: x.interpolate())

print("remaining missing after interpolation:", master_model_df["price_lcu_per_tonne"].isnull().sum())
print(master_model_df[["crop_name", "year", "price_lcu_per_tonne"]].head(20))
```

tep 1: fix real price

You already fixed nominal price, so now recompute:
1. Adjusting prices for inflation

You already have:

price_lcu_per_tonne in local currency
food_cpi

So after merging, create a real price column:

```python
master_model_df["real_price_lcu_per_tonne"] = (
    master_model_df["price_lcu_per_tonne"] / (master_model_df["food_cpi"] / 100)
)

print("missing real price:", master_model_df["real_price_lcu_per_tonne"].isnull().sum())
```

2. Macro variables in USD

That is also okay.

Your macro variables are:

gdp_usd_2015
agriculture_value_added_usd_2015

These are already in constant 2015 USD, which is actually good because:

they are inflation-adjusted already
they are comparable across years

So you do not need to convert them unless you specifically want all money variables in

```python
master_model_df["gdp_kes_2015_equiv"] = (
    master_model_df["gdp_usd_2015"] * master_model_df["kes_per_usd"]
)

master_model_df["agriculture_value_added_kes_2015_equiv"] = (
    master_model_df["agriculture_value_added_usd_2015"] * master_model_df["kes_per_usd"]
)
```
