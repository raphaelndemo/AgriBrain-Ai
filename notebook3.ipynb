🧠 Case 1: crops with full price history (ideal)

Example:

maize
rice
wheat

👉 These have prices across most or all years
✔ best for modelling

⚠️ Case 2: crops with partial price data

Example:

apples
onions
tomatoes

👉 These:

may have prices only from, say, 2015–2024
earlier years are missing

✔ still usable
BUT need interpolation (which we already planned)

❌ Case 3: crops with no price at all

These are the ones causing your 307 missing values.

Example:

cloves
palm kernels
oats

👉 These:

never appear in producer prices
are pure noise for price modelling

```python
price_coverage = master_df.groupby("crop_name")["price_lcu_per_tonne"].agg(
    total_rows="size",
    non_null_prices=lambda x: x.notnull().sum(),
    percent_available=lambda x: x.notnull().mean() * 100
).reset_index()

print(price_coverage.sort_values("percent_available"))
```

Step: Filter dataset properly

```python
good_crops = price_coverage[
    price_coverage["percent_available"] >= 50
]["crop_name"].tolist()

master_model_df = master_df[
    master_df["crop_name"].isin(good_crops)
].copy()

print("final crops:", len(good_crops))
print(sorted(good_crops))

print("new shape:", master_model_df.shape)
print("remaining missing prices:", master_model_df["price_lcu_per_tonne"].isnull().sum())
```

Next step: interpolate prices within each crop

```python
master_model_df = master_model_df.sort_values(["crop_name", "year"]).copy()

master_model_df["price_lcu_per_tonne"] = master_model_df.groupby("crop_name")[
    "price_lcu_per_tonne"
].transform(lambda x: x.interpolate())

print("remaining missing after interpolation:", master_model_df["price_lcu_per_tonne"].isnull().sum())
print(master_model_df[["crop_name", "year", "price_lcu_per_tonne"]].head(20))
```

Merging KAMIS and Faostat data

Layer 1 = FAOSTAT annual structural layer
This explains the long-run agricultural story, i.e. production, imports, exports, supply, demand, balance, prices, macro conditions.


Layer 2 = KAMIS short-term market layer
This explains what is happening now in real markets, i.e. daily or near-daily wholesale, retail, supply volumes, county/market movement.

That is better than forcing both datasets into one shape too early.

These two datasets are not at the same grain.


And when two datasets are at different grains, forcing a direct merge usually creates messy duplication and misleading rows.


# why other options are weaker
option 1: fully aggregate KAMIS and merge everything into annual only

This can work for a first baseline model, but it weakens the project if used as the only architecture.

Why it is limited:

you lose the live or near-live market pulse
you flatten daily variation into yearly summaries
the system becomes good at historical structure, but weaker for current market advice
AgriBrain is meant to warn farmers about timing and market glut risk, and daily or monthly KAMIS helps with that

So this option is acceptable for a simple modelling notebook, but not ideal for the full product vision.

option 2: force everything into monthly or daily

This sounds attractive, but it creates new problems.

Why it is difficult:

your FAOSTAT data is annual, not daily
yearly values like GDP, agriculture value added, fertilizer, and food balance would get repeated across many monthly rows
that repetition can make the model think it has more information than it actually does
some annual variables are not truly monthly signals, so repeating them can distort interpretation
you also increase data size and complexity before you have a stable baseline

So this option gives more granularity, but at the cost of more assumptions and more modelling risk.

option 3: two-layer design

This is the strongest design because it lets each dataset do the job it is best at.

FAOSTAT layer answers:
what is the historical supply-demand structure?
which crops tend to be in shortage or oversupply over time?
how do macro conditions and farm-level prices relate to that pattern?
KAMIS layer answers:
what is happening in markets now?
which counties or markets are seeing price spikes?
which commodities are volatile this month?
where is supply tightening or increasing?

This means your system can later combine:

structural intelligence from FAOSTAT
short-term intelligence from KAMIS

That is exactly the kind of split that makes sense for AgriBrain.

what we will do now

Before feature engineering and before XGBoost, we need to navigate both datasets properly.

So our first coding step is not merging yet.

It is:

code block 2: clean KAMIS

```python
# -----------------------------
# STEP 2: CLEAN KAMIS DATA
# -----------------------------

kamis_clean = kamis_layer2.copy()

# -----------------------------
# 1. DROP USELESS COLUMN
# -----------------------------
# 'Unnamed: 0' is just an index from CSV export
kamis_clean = kamis_clean.drop(columns=['Unnamed: 0'])

# -----------------------------
# 2. REMOVE DUPLICATE COLUMN
# -----------------------------
# 'Commodity' and 'commodity' are duplicates
# keep one (standardize to lowercase later)
kamis_clean = kamis_clean.drop(columns=['commodity'])

# -----------------------------
# 3. STANDARDIZE COLUMN NAMES
# -----------------------------
kamis_clean.columns = [
    'commodity', 'market', 'wholesale_price',
    'retail_price', 'supply_volume',
    'county', 'date', 'commodity_id'
]

# -----------------------------
# 4. CONVERT DATE TO DATETIME
# -----------------------------
kamis_clean['date'] = pd.to_datetime(kamis_clean['date'])

# Extract year and month (important later)
kamis_clean['year'] = kamis_clean['date'].dt.year
kamis_clean['month'] = kamis_clean['date'].dt.month

# -----------------------------
# 5. HANDLE ZERO PRICES
# -----------------------------
# In market data, 0 often means missing, not real price

kamis_clean['wholesale_price'] = kamis_clean['wholesale_price'].replace(0, pd.NA)
kamis_clean['retail_price'] = kamis_clean['retail_price'].replace(0, pd.NA)

# -----------------------------
# 6. QUICK CHECK
# -----------------------------
print("Shape after cleaning:", kamis_clean.shape)

print("\nColumns:")
print(kamis_clean.columns)

print("\nDate range:")
print(kamis_clean['date'].min(), "to", kamis_clean['date'].max())

print("\nMissing values:")
print(kamis_clean.isna().sum())

print("\nSample:")
print(kamis_clean.head())
```

```python
"""
Step 3: design the aggregation (this is where your model quality is decided)
🧠 first, interpret what you just got

From your output:

good news
Data spans 2005 → 2026 → aligns well with FAOSTAT ✔️
You now have year + month → enables flexible aggregation ✔️
Clean structure → ready for grouping ✔️
important observations
1. missing prices
~82k missing wholesale
~51k missing retail

👉 this is normal in market data
👉 we will not drop rows, we will handle this smartly

2. supply_volume has no missing

👉 very valuable → we can use it as:

weighting factor
market activity signal
3. multiple markets per day

Example:

Dry Maize | Nairobi | 2026-04-17
Dry Maize | Kakamega | 2026-04-17

👉 This means:

we have spatial variation
we must collapse it carefully

🎯 now the key question: how do we aggregate?

You were worried about losing intelligence.
So we will NOT just take averages.

We will build market behaviour features.
"""
```

```python
"""
step 3: aggregation design (concept first)
🔴 what we MUST preserve

From KAMIS, we want to capture:

1. price level
average price
2. price volatility
how unstable prices are
3. price extremes
max / min
4. market spread
retail vs wholesale difference....So we will create:
-avg_wholesale_price
-avg_retail_price
-price_spread
5. supply signal
how much is flowing in markets
"""
```

Fix numeric dtypes in wholwsale, retail and supply columns first

```python
# -----------------------------
# STEP 3A: FIX NUMERIC TYPES
# -----------------------------
# After replacing 0 with pd.NA, pandas may treat the columns as object type.
# We convert them back to numeric so aggregation functions like mean/std work.

numeric_cols = ['wholesale_price', 'retail_price', 'supply_volume']

for col in numeric_cols:
    kamis_clean[col] = pd.to_numeric(kamis_clean[col], errors='coerce')

print(kamis_clean[numeric_cols].dtypes)
print("\nMissing values after numeric conversion:")
print(kamis_clean[numeric_cols].isna().sum())
```

```python
# -----------------------------
# STEP 3B: AGGREGATE KAMIS TO YEARLY FEATURES
# -----------------------------
# We convert high-frequency market observations into yearly market-intelligence
# features that can later be aligned with the FAOSTAT annual structure.

kamis_yearly = (
    kamis_clean
    .groupby(['commodity', 'year'])
    .agg(
        avg_wholesale_price=('wholesale_price', 'mean'),
        std_wholesale_price=('wholesale_price', 'std'),
        max_wholesale_price=('wholesale_price', 'max'),
        min_wholesale_price=('wholesale_price', 'min'),

        avg_retail_price=('retail_price', 'mean'),
        std_retail_price=('retail_price', 'std'),
        max_retail_price=('retail_price', 'max'),
        min_retail_price=('retail_price', 'min'),

        total_supply_volume=('supply_volume', 'sum'),
        avg_supply_volume=('supply_volume', 'mean'),
        market_count=('market', 'nunique'),
        county_count=('county', 'nunique'),
        observation_count=('date', 'count')
    )
    .reset_index()
)

# -----------------------------
# ADD DERIVED MARKET FEATURES
# -----------------------------
# These preserve more intelligence than a simple average.

kamis_yearly['price_spread'] = (
    kamis_yearly['avg_retail_price'] - kamis_yearly['avg_wholesale_price']
)

kamis_yearly['retail_price_range'] = (
    kamis_yearly['max_retail_price'] - kamis_yearly['min_retail_price']
)

kamis_yearly['wholesale_price_range'] = (
    kamis_yearly['max_wholesale_price'] - kamis_yearly['min_wholesale_price']
)

print("Shape:", kamis_yearly.shape)
print("\nColumns:")
print(kamis_yearly.columns.tolist())
print("\nSample:")
print(kamis_yearly.head())
```

```python
"""
why this version is better than the earlier one

This version uses named aggregation, which is cleaner and safer.

Instead of writing a dictionary with repeated column keys, we explicitly say:

avg_wholesale_price=('wholesale_price', 'mean')

That avoids overwriting issues and gives clean column names immediately.

what these features mean

This step is important because you said you want a fuller explanation of what we are doing and why.

avg_wholesale_price and avg_retail_price

These capture the general annual price level of the commodity across markets.

Why useful:

high average prices can indicate tight supply
low average prices can indicate glut or oversupply
std_wholesale_price and std_retail_price

These capture price volatility.

Why useful:

volatile prices often signal unstable markets
stable prices suggest more predictable supply conditions

This is one of the ways we preserve market intelligence instead of destroying it.

max_* and min_*

These capture price extremes.

Why useful:

a high max price may capture a scarcity spike
a low min price may capture glut or harvest-time saturation

So even though we are aggregating, we are still preserving within-year stress signals.

total_supply_volume and avg_supply_volume

These represent market flow.

Why useful:

high total supply can reflect strong availability
weak supply can indicate scarcity

Since supply volume had no missing values in your KAMIS output, this is a particularly valuable feature.

market_count and county_count

These capture coverage and spread.

Why useful:

a commodity appearing in many markets or counties may indicate broad market presence
limited spread may indicate localized availability

These are not just technical counts. They can reflect commodity reach in the market system.

observation_count

This tells us how many records support the yearly summary.

Why useful:

helps judge how reliable the yearly aggregate is
low-count years may need caution later
price_spread

This is:

avg retail price - avg wholesale price

Why useful:

reflects margin between wholesale and consumer prices
can indicate transport costs, intermediation, inefficiency, or bottlenecks

This is very useful for AgriBrain because it gives more than just a price level. It gives market structure.

why we are doing this instead of keeping raw daily rows

Because your FAOSTAT table is annual.

If we merge raw daily KAMIS directly into annual FAOSTAT:

one FAOSTAT row becomes many duplicated rows
structural variables repeat thousands of times
the modelling table becomes distorted
the model may over-weight repeated annual values

So aggregation here is not a loss of purpose. It is a translation step from market-frequency data into annual market intelligence features.

That is why the two-layer idea works:

raw KAMIS still exists as a short-term market layer
aggregated KAMIS becomes model-ready for the annual structural layer

So you are keeping both:

detailed market data for later app logic
yearly market features for XGBoost training
"""
```

🎯 STEP 7: clean extreme values (very important)

We will cap unrealistic values (simple and effective

```python

# STEP 7: HANDLE OUTLIERS

kamis_final_clean = kamis_final.copy()

# Cap extreme values using quantiles (robust method)

for col in [
    'avg_wholesale_price', 'avg_retail_price',
    'max_wholesale_price', 'max_retail_price',
    'min_wholesale_price', 'min_retail_price'
]:
    lower = kamis_final_clean[col].quantile(0.01)
    upper = kamis_final_clean[col].quantile(0.99)

    kamis_final_clean[col] = kamis_final_clean[col].clip(lower, upper)

# Check again
print(kamis_final_clean.describe())
```

🧠 goal of this step

We want to:

remove extreme noise (0.01, 40,000 etc.)
keep real market variation
NOT distort the data

👉 So we use quantile capping (winsorization)

```python

# STEP 7: HANDLE OUTLIERS (CAPPING)

kamis_final_clean = kamis_final.copy()

# Columns to clean (price-related only)
price_cols = [
    'avg_wholesale_price', 'std_wholesale_price',
    'max_wholesale_price', 'min_wholesale_price',
    'avg_retail_price', 'std_retail_price',
    'max_retail_price', 'min_retail_price',
    'price_spread',
    'retail_price_range', 'wholesale_price_range'
]

# Apply capping using 1st and 99th percentiles
for col in price_cols:
    lower = kamis_final_clean[col].quantile(0.01)
    upper = kamis_final_clean[col].quantile(0.99)

    kamis_final_clean[col] = kamis_final_clean[col].clip(lower, upper)

# CHECK RESULTS
print(kamis_final_clean[price_cols].describe())
```

STEP 7B: remove unrealistic minimums (critical fix)

```python

# STEP 7B: FIX UNREALISTIC MIN VALUES


kamis_final_clean2 = kamis_final_clean.copy()

# Replace very small values (below 1) with NaN
for col in [
    'min_wholesale_price',
    'min_retail_price'
]:
    kamis_final_clean2[col] = kamis_final_clean2[col].apply(
        lambda x: None if x < 1 else x
    )

# Recalculate price ranges (important after cleaning)
kamis_final_clean2['retail_price_range'] = (
    kamis_final_clean2['max_retail_price'] - kamis_final_clean2['min_retail_price']
)

kamis_final_clean2['wholesale_price_range'] = (
    kamis_final_clean2['max_wholesale_price'] - kamis_final_clean2['min_wholesale_price']
)

# Check again
print(kamis_final_clean2.describe())
```

```python
# -----------------------------
# STEP 10A: FEATURE ENGINEERING (CORE SIGNALS)
# -----------------------------

df = final_model.copy()

# -----------------------------
# 1. SUPPLY-DEMAND SIGNALS
# -----------------------------
df['supply_demand_ratio'] = df['supply'] / (df['demand'] + 1e-6)
df['net_trade'] = df['import_quantity'] - df['export_quantity']

# -----------------------------
# 2. PRICE SIGNALS
# -----------------------------
df['real_price'] = df['price_lcu_per_tonne'] / (df['food_cpi'] + 1e-6)
df['market_price_gap'] = df['avg_retail_price'] - df['price_lcu_per_tonne']

# -----------------------------
# 3. INPUT INTENSITY
# -----------------------------
df['fertilizer_total'] = df['nitrogen'] + df['phosphorus'] + df['potassium']
df['pesticide_intensity'] = df['pesticides_total']

# -----------------------------
# 4. MACRO CONTEXT
# -----------------------------
df['agri_share'] = df['agriculture_value_added_usd_2015'] / (df['gdp_usd_2015'] + 1e-6)

# -----------------------------
# CHECK NEW FEATURES
# -----------------------------
print(df[['supply_demand_ratio', 'net_trade',
          'real_price', 'market_price_gap',
          'fertilizer_total', 'agri_share']].head())
```
